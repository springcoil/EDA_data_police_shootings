```python
data = df[['manner_of_death', 'armed', 'race']]
data
```

```python

table = pd.pivot_table(data, index='armed',columns='race', values='manner_of_death', aggfunc=np.count_nonzero)
```

```python
table = pd.pivot_table(data,columns='race', values='manner_of_death', aggfunc=np.count_nonzero)
df_data_race_manner_of_death = pd.DataFrame(table).reset_index()
```

```python
id_vars = ["date",
           "manner_of_death",
           "armed",
           "age"]

df_melted = pd.melt(frame=df,id_vars=id_vars, value_vars='race')
```
